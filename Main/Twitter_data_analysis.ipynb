# Twitter Data Analysis

```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
```

```python
query1 = "(#travelindia, OR #exploreindia, OR #TravelIndia, OR #PlacesofIndia, OR #unknownindia, OR #exploringindia, OR #beautifulindia, OR #indiatourism, OR #andhratourism, OR #haryanatourism, OR #travelharyana, OR #manipurtourism, OR #travelmanipur, OR #sikkimtourism, OR #arunachaltourism, OR #himachaltourism, OR #TravelHimachalPradesh, OR #meghalayatourism, OR #TamilNaduTourism, OR #AssamTourism, OR #TravelAssam, OR #Jharkhandtourism, OR #mizoramtourism, OR #TelanganaTourism, OR #BiharTourism)"

query2 = "(#KarnatakaTourism, OR #TravelKarnataka, OR #NagalandTourism, OR #NortheastIndiaTourism, OR #TripuraTourism, OR #VisitTripura, OR #VisitHimachal, OR #ChattisgarhTourism, OR #KeralaTourism, OR #VisitKerala, OR #TravelKerala, OR #OdishaTourism, OR #VisitOdisha, OR #UttrakhandTourism, OR #TravelUttrakhand, OR #visitUttrakhand, OR #GoaTourism, OR #VisitGoa, OR #TravelGoa, OR #MadhyaPradeshTourism, OR #PunjabTourism, OR #VisitPunjab, OR #UttarPradeshTourism, OR #UPTourism, OR #VisitUttarPradesh)"

query3 = "(#GujratTourism, OR #VisitGujrat, OR #ExploreGujrat, OR #TravelGujrat, OR #MaharashtraTourism, OR #VisitMaharashtra, OR #TravelMaharashtra, OR #MumbaiTourism, OR #RajasthanTourism, OR #VisitRajathan, OR #ExploreRajasthan, OR #TouristPlacesOfRajasthan, OR #WestBengalTourism, OR #KolkataTourism, OR #AndamanTourism, OR #TravelAndamanAndNicobar, OR #DadraandNagarHaveliTourism, OR #DamanandDiuTourism, OR #JammuTourism, OR #KashmirTourism, OR #ExploreKashmir, OR #LakshadweepTourism)"

query4 = "(#VisitKashmir, OR #NewDelhiTOurism, OR #TourDelhi, OR #ExploreDelhi, OR #VisitDelhi, OR #PuducherryTourism, OR #LehTourism, OR #ChandigarhTOurism, OR #LadakhTourism)"

```

```python
tweets_query1 =[]
limit = 5000
for tweet in sntwitter.TwitterHashtagScraper(query1).get_items():
#     print(vars(tweet))
#     break
    if len(tweets_query1) == limit:
        break
        
    else:
        tweets_query1.append([tweet.date, tweet.user.username,tweet.place ,tweet.coordinates, tweet.rawContent])
        
df_1 = pd.DataFrame(tweets_query1, columns=['Date', 'Username', 'Place', 'Coordinates','Contents'])
    
```

```python
final_df['Coordinates'] = final_df['Coordinates'].astype(str)

# extract longitude and latitude values from the column
final_df['longitude'] = final_df['Coordinates'].str.split('longitude=').str[1].str.split(',').str[0].astype(float)
final_df['latitude'] = final_df['Coordinates'].str.split('latitude=').str[1].str.split(')').str[0].astype(float)


# drop the original column
final_df.drop('Coordinates', axis=1, inplace=True)

# save the updated dataframe to a new csv file
final_df.to_csv('twitter_posts.csv', index=False)
```

```python
final_df['Place'] = final_df['Place'].astype(str)

# extract fullName and country values from the column
final_df['city'] = final_df['Place'].str.split("fullName='").str[1].str.split(',').str[0]
final_df['country'] = final_df['Place'].str.split("country='").str[1].str.split('\'').str[0]

# drop the original column
final_df.drop('Place', axis=1, inplace=True)

# save the updated dataframe to a new csv file
final_df.to_csv('twitter_posts.csv', index=False)
```
