#####  Analyse the Contents column for the datapoints with no lat, long mentioned

Find out name, place and organisation that can be extracted from the Contetnt using the NLP method used in previous project.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib import style
from nltk import pos_tag
from nltk.tag import StanfordNERTagger
from nltk.tokenize import word_tokenize

style.use('fivethirtyeight')
```

```python
df = pd.read_csv('final_twitter_posts.csv', lineterminator='\n')
```

```python
df.head(10)
```

```python
np.savetxt(r'Contents.txt', df['Contents'].values, fmt='%s')
```

```python
def process_text(txt_file):
    raw_text = open("Contents.txt").read()
    token_text = word_tokenize(raw_text)
    return token_text
```

```python
def stanford_tagger(token_text):
    st = StanfordNERTagger('stanford-ner-2020-11-17/classifiers/english.all.3class.distsim.crf.ser.gz','stanford-ner-2020-11-17/stanford-ner.jar', encoding='utf-8')   
    ne_tagged = st.tag(token_text)
    locations = [word for word, tag in ne_tagged if tag == 'LOCATION']
    np.savetxt("Locations.csv", locations, delimiter=", ", fmt="%s")
    print(locations)
    return(ne_tagged)
```

```python
def nltk_tagger(token_text):
    tagged_words = nltk.pos_tag(token_text)
    ne_tagged = nltk.ne_chunk(tagged_words)
    return(ne_tagged)
```

```python
def stanford_main():
    txt_file = open("Contents.txt").read()
    stanford_tagger(process_text(txt_file))
    
```

```python
stanford_main()
```
